 ## Example 0: (bond-graph model from nxgraph)
This example shows how to obtain BG model from nx graph.

```python
import matplotlib.pyplot as plt
import sys
import numpy as np

import sympy as sp
import re

sys.path.append("../run/")
import BondGraph as bg
import utils

import networkx as nx
import grammar as grammar

import random

from tqdm import tqdm
```

```python
graph = grammar.visualize_with_ports_and_subgraphs()
```

```python
# Создаём файл для записи отладочной информации
debug_file = open("debug_output_for_demo1.txt", "w")

# Функция для записи отладочной информации
def log_debug_info(message):
    print(message)  # Вывод в консоль
    debug_file.write(message + "\n")  # Запись в файл

# Создаём модель BondGraph
model = bg.BondGraph(Name='Generated Bond Graph')
model.debug = False

# Словарь для хранения созданных элементов BondGraph
bondgraph_elements = {}

# Функция для сопоставления типов узлов с элементами BondGraph
def create_bondgraph_element(node_id, node_type):
    # Убираем символ '*' из типа узла
    element_type = node_type.replace('*', '')
    bondgraph_element = bg.BGelement(element_type)  # Убираем аргумент ID
    bondgraph_elements[node_id] = bondgraph_element
    return bondgraph_element

# Преобразуем узлы графа в элементы BondGraph
for node, data in graph.nodes(data=True):
    node_id = node
    node_type = data["type"]
    element = create_bondgraph_element(node_id, node_type)
    model.addElement(element)

# Добавляем соединения на основе рёбер графа
for src, dst in graph.edges():
    if src in bondgraph_elements and dst in bondgraph_elements:
        model.connect(bondgraph_elements[src], bondgraph_elements[dst])

# Применяем правила, визуализируем и проверяем модель
model.applyRules()
model.render()
model.draw()

# Показать порты и уравнения
bg.show_ports_state(model)

# Записываем информацию об элементах и их портах
log_debug_info("Element and port details:")
for element in model.getElementsList():
    log_debug_info(f"Element name:  {element.getName()}")
    for port in element.getPorts():
        log_debug_info(
            f"++ port_id: {port.getId()} |  arrow:  {port.getDirection()} |  causality:  {port.getCausality()}"
        )

# Проверяем наличие ошибок
log_debug_info("\nChecking bonds for missing efforts or flows:")
for bond in model.getBondList():
    if not bond.effort or not bond.flow:
        log_debug_info(f"ERROR 007: The bond id: {bond.getId()}  has no effort or flow")


# Выводим уравнения в консоль
model.assign_equations()
equations = model.get_model_equations() # if you want to get the system of equations
log_debug_info("\n Model equations:")
for eq in equations:
    log_debug_info(eq if eq else "Empty equation")

    
equations
# Закрываем файл
debug_file.close()
```

### You can just crete variables with elements and attach them later
As shown below

```python
#Create some elements (this is unique elements with unique ID, hense if you reinitialize th sam variable you will get different component inside)
SE_1 = bg.BGelement('SE')
SF_1 = bg.BGelement('SF', Position=[2,1]) # Position is only for visualization so you can skip and it will be assign automatically 
C_1 = bg.BGelement('C', Position=[3,6])
R_1 = bg.BGelement('R', Position=[3,4])
I_1 = bg.BGelement('I', Position=[0,3])
OneJ_1 = bg.BGelement('1', Position=[0,1])
OneJ_2 = bg.BGelement('1', Position=[1,1])
OneJ_3 = bg.BGelement('1', Position=[2,0])
ZeroJ_1 = bg.BGelement('0', Position=[1,0])
```

```python

initial_state = np.zeros((len(model.state_variables),1)) # fill initials by zero
print("initial_state: ", initial_state)
time_steps=500
sampling_period=0.05
input_sequence =np.ones((time_steps, len(model.input_variables)))
# print("Input sequence:", input_sequence)
input_sequence[:, 0] = 1
```

```python
state, output = model.simulate(initial_state, input_sequence, time_steps, sampling_period, parameters_values)    
```
